```python
import requests
import os

try:
    from IPython.display import Audio, display
    ipython_available = True
except ImportError:
    ipython_available = False

# --- Cáº¤U HÃŒNH ---
NGROK_URL = "https://347b824096e0.ngrok-free.app"

# VÄƒn báº£n báº¡n muá»‘n chuyá»ƒn thÃ nh giá»ng nÃ³i
TEXT_TO_SYNTHESIZE = "QuÃ£ng thá»i gian Ä‘i há»c lÃ  má»™t trong nhá»¯ng kÃ½ á»©c Ä‘áº¹p nháº¥t trong cuá»™c Ä‘á»i má»—i con ngÆ°á»i. Má»—i mÃ¹a khai giáº£ng lÃ  má»™t dáº¥u má»‘c Ä‘áº·c biá»‡t, luÃ´n lÃ  dá»‹p Ä‘Æ°á»£c nhiá»u sinh viÃªn mong Ä‘á»£i. HÃ ng nÄƒm, cá»© vÃ o dá»‹p thÃ¡ng 9, 10 lÃ  sinh viÃªn Äáº¡i há»c BÃ¡ch khoa HÃ  Ná»™i láº¡i nÃ´ ná»©c tham dá»± Lá»… khai giáº£ng, mang theo niá»m vui, hi vá»ng vá» má»™t nÄƒm há»c má»›i. NgÃ y khai giáº£ng nÄƒm há»c má»›i luÃ´n ráº¥t Ã½ nghÄ©a vÃ  Ä‘Æ°á»£c mong Ä‘á»£i Ä‘á»‘i vá»›i cÃ¡c em tÃ¢n sinh viÃªn."

# TÃªn file Ä‘á»ƒ lÆ°u Ã¢m thanh Ä‘áº§u ra trÃªn mÃ¡y cá»§a báº¡n
OUTPUT_FILENAME = "downloaded_speech_from_colab.wav"

# --- KIá»‚M TRA VÃ€ Gá»¬I YÃŠU Cáº¦U ---
if NGROK_URL is None:
    print("Update ngrok url.")
else:
    # Endpoint cho chá»©c nÄƒng Text-to-Speech
    api_endpoint = f"{NGROK_URL}/tts"
    
    # Dá»¯ liá»‡u gá»­i Ä‘i trong body cá»§a request POST
    payload = {"text": TEXT_TO_SYNTHESIZE, "speaker_id" : 1}

    print(f"â–¶ï¸ Äang gá»­i yÃªu cáº§u POST tá»›i: {api_endpoint}")
    print(f"   VÄƒn báº£n: '{TEXT_TO_SYNTHESIZE}'")

    try:
        # Gá»­i yÃªu cáº§u POST tá»›i endpoint /tts vá»›i payload lÃ  JSON vÃ  tÄƒng thá»i gian chá»
        response = requests.post(api_endpoint, json=payload, timeout=60)

        # Kiá»ƒm tra xem yÃªu cáº§u cÃ³ thÃ nh cÃ´ng khÃ´ng
        if response.status_code == 200:
            # Láº¥y ná»™i dung (dá»¯ liá»‡u bytes cá»§a file audio)
            audio_bytes = response.content

            # Ghi dá»¯ liá»‡u bytes vÃ o file
            with open(OUTPUT_FILENAME, 'wb') as f:
                f.write(audio_bytes)

            print(f"\nâœ… Tá»•ng há»£p vÃ  táº£i vá» thÃ nh cÃ´ng!")
            print(f"   ğŸ’¾ File Ã¢m thanh Ä‘Ã£ Ä‘Æ°á»£c lÆ°u táº¡i: {os.path.abspath(OUTPUT_FILENAME)}")
            print(f"   ğŸ“¦ KÃ­ch thÆ°á»›c file: {len(audio_bytes) / 1024:.2f} KB")

            # Náº¿u cÃ³ thá»ƒ, phÃ¡t Ã¢m thanh trá»±c tiáº¿p
            if ipython_available:
                print("\nğŸ”Š Äang phÃ¡t Ã¢m thanh trá»±c tiáº¿p...")
                display(Audio(data=audio_bytes, autoplay=True))

        else:
            print(f"\nâŒ Lá»—i! Server tráº£ vá» mÃ£ tráº¡ng thÃ¡i: {response.status_code}")
            print(f"   Ná»™i dung pháº£n há»“i: {response.text}")

    except requests.exceptions.RequestException as e:
        print(f"\nâŒ ÄÃ£ xáº£y ra lá»—i káº¿t ná»‘i: {e}")
        print("   Vui lÃ²ng kiá»ƒm tra láº¡i nhá»¯ng Ä‘iá»u sau:")
        print("     1. URL ngrok Ä‘Ã£ chÃ­nh xÃ¡c chÆ°a.")
        print("     2. Server trÃªn Colab cÃ³ cÃ²n Ä‘ang cháº¡y khÃ´ng.")
        print("     3. Káº¿t ná»‘i máº¡ng cá»§a báº¡n cÃ³ á»•n Ä‘á»‹nh khÃ´ng.")

```

```python
# YÃªu cáº§u cÃ i Ä‘áº·t thÆ° viá»‡n: pip install gTTS ipython
import os
import io
from gtts import gTTS

# ThÆ° viá»‡n Ä‘á»ƒ phÃ¡t audio trá»±c tiáº¿p trong Jupyter Notebook/Lab
try:
    from IPython.display import Audio, display
    ipython_available = True
except ImportError:
    ipython_available = False

# --- Cáº¤U HÃŒNH ---
# VÄƒn báº£n báº¡n muá»‘n chuyá»ƒn thÃ nh giá»ng nÃ³i
TEXT_TO_SYNTHESIZE = "QuÃ£ng thá»i gian Ä‘i há»c lÃ  má»™t trong nhá»¯ng kÃ½ á»©c Ä‘áº¹p nháº¥t trong cuá»™c Ä‘á»i má»—i con ngÆ°á»i. Má»—i mÃ¹a khai giáº£ng lÃ  má»™t dáº¥u má»‘c Ä‘áº·c biá»‡t, luÃ´n lÃ  dá»‹p Ä‘Æ°á»£c nhiá»u sinh viÃªn mong Ä‘á»£i. HÃ ng nÄƒm, cá»© vÃ o dá»‹p thÃ¡ng 9, 10 lÃ  sinh viÃªn Äáº¡i há»c BÃ¡ch khoa HÃ  Ná»™i láº¡i nÃ´ ná»©c tham dá»± Lá»… khai giáº£ng, mang theo niá»m vui, hi vá»ng vá» má»™t nÄƒm há»c má»›i. NgÃ y khai giáº£ng nÄƒm há»c má»›i luÃ´n ráº¥t Ã½ nghÄ©a vÃ  Ä‘Æ°á»£c mong Ä‘á»£i Ä‘á»‘i vá»›i cÃ¡c em tÃ¢n sinh viÃªn."

# TÃªn file Ä‘á»ƒ lÆ°u Ã¢m thanh Ä‘áº§u ra trÃªn mÃ¡y cá»§a báº¡n
OUTPUT_FILENAME = "gtts_speech_output.mp3" # gTTS thÆ°á»ng xuáº¥t ra file .mp3

# --- Tá»”NG Há»¢P VÃ€ LÆ¯U FILE ---
print(f"â–¶ï¸ Äang tá»•ng há»£p vÄƒn báº£n báº±ng gTTS Ğ»Ğ¾ĞºĞ°Ğ»ÑŒĞ½Ğ¾...")
print(f"   VÄƒn báº£n: '{TEXT_TO_SYNTHESIZE}'")

try:
    # Táº¡o Ä‘á»‘i tÆ°á»£ng gTTS vá»›i vÄƒn báº£n vÃ  ngÃ´n ngá»¯ tiáº¿ng Viá»‡t ('vi')
    tts = gTTS(text=TEXT_TO_SYNTHESIZE, lang='vi', slow = False)

    # Sá»­ dá»¥ng má»™t Ä‘á»‘i tÆ°á»£ng bytes trong bá»™ nhá»› Ä‘á»ƒ lÆ°u file audio táº¡m thá»i
    audio_fp = io.BytesIO()
    tts.write_to_fp(audio_fp)
    # ÄÆ°a con trá» vá» Ä‘áº§u file Ä‘á»ƒ Ä‘á»c láº¡i
    audio_fp.seek(0)
    audio_bytes = audio_fp.read()

    # Ghi dá»¯ liá»‡u bytes vÃ o file trÃªn Ä‘Ä©a
    with open(OUTPUT_FILENAME, 'wb') as f:
        f.write(audio_bytes)

    print(f"\nâœ… Tá»•ng há»£p thÃ nh cÃ´ng!")
    print(f"   ğŸ’¾ File Ã¢m thanh Ä‘Ã£ Ä‘Æ°á»£c lÆ°u táº¡i: {os.path.abspath(OUTPUT_FILENAME)}")
    print(f"   ğŸ“¦ KÃ­ch thÆ°á»›c file: {len(audio_bytes) / 1024:.2f} KB")

    # Náº¿u cÃ³ thá»ƒ, phÃ¡t Ã¢m thanh trá»±c tiáº¿p
    if ipython_available:
        print("\nğŸ”Š Äang phÃ¡t Ã¢m thanh trá»±c tiáº¿p...")
        # autoplay=True Ä‘á»ƒ tá»± Ä‘á»™ng phÃ¡t
        display(Audio(data=audio_bytes, autoplay=True))

except Exception as e:
    print(f"\nâŒ ÄÃ£ xáº£y ra lá»—i khi sá»­ dá»¥ng gTTS: {e}")
    print("   Vui lÃ²ng kiá»ƒm tra láº¡i nhá»¯ng Ä‘iá»u sau:")
    print("     1. Káº¿t ná»‘i máº¡ng cá»§a báº¡n cÃ³ á»•n Ä‘á»‹nh khÃ´ng (gTTS cáº§n internet).")
    print("     2. ThÆ° viá»‡n gTTS Ä‘Ã£ Ä‘Æ°á»£c cÃ i Ä‘áº·t Ä‘Ãºng cÃ¡ch chÆ°a.")

```

```python
import requests
import os

# --- Cáº¤U HÃŒNH ---
# Äá»‹a chá»‰ server backend cá»§a báº¡n
API_BASE_URL = "http://127.0.0.1:8000"

def test_ask_endpoint():
    """
    Kiá»ƒm tra endpoint /ask Ä‘á»ƒ nháº­n cÃ¢u tráº£ lá»i tá»« chatbot.
    """
    print("--- Báº¯t Ä‘áº§u kiá»ƒm tra endpoint: /ask ---")
    url = f"{API_BASE_URL}/ask"
    question = "Há»c bá»•ng dÃ nh cho sinh viÃªn cÃ³ hoÃ n cáº£nh khÃ³ khÄƒn?"
    payload = {"question": question}

    try:
        print(f"Gá»­i cÃ¢u há»i: '{question}'")
        response = requests.post(url, json=payload, timeout=30) # timeout 30 giÃ¢y

        # Kiá»ƒm tra xem yÃªu cáº§u cÃ³ thÃ nh cÃ´ng khÃ´ng (status code 2xx)
        response.raise_for_status()

        data = response.json()
        print(f"âœ… [THÃ€NH CÃ”NG] Server pháº£n há»“i:")
        print(f"   CÃ¢u tráº£ lá»i: {data.get('answer', 'KhÃ´ng cÃ³ cÃ¢u tráº£ lá»i')[:100]}...") # In 100 kÃ½ tá»± Ä‘áº§u

    except requests.exceptions.RequestException as e:
        print(f"âŒ [THáº¤T Báº I] ÄÃ£ xáº£y ra lá»—i khi gá»i API: {e}")
    except Exception as e:
        print(f"âŒ [THáº¤T Báº I] Lá»—i khÃ´ng xÃ¡c Ä‘á»‹nh: {e}")
    finally:
        print("--- Káº¿t thÃºc kiá»ƒm tra endpoint: /ask ---\n")


def test_tts_endpoint():
    """
    Kiá»ƒm tra endpoint /tts Ä‘á»ƒ chuyá»ƒn vÄƒn báº£n thÃ nh giá»ng nÃ³i.
    """
    print("--- Báº¯t Ä‘áº§u kiá»ƒm tra endpoint: /tts ---")
    url = f"{API_BASE_URL}/tts"
    text_to_speak = "Xin chÃ o, Ä‘Ã¢y lÃ  giá»ng nÃ³i Ä‘Æ°á»£c táº¡o ra tá»« server."
    payload = {"text": text_to_speak}
    output_filename = "backend_test_output.mp3"

    try:
        print(f"Gá»­i vÄƒn báº£n: '{text_to_speak}'")
        response = requests.post(url, json=payload, timeout=20)

        response.raise_for_status()

        # Kiá»ƒm tra header Ä‘á»ƒ cháº¯c cháº¯n nháº­n vá» file Ã¢m thanh
        if 'audio/mpeg' in response.headers.get('Content-Type', ''):
            # LÆ°u file Ã¢m thanh nháº­n Ä‘Æ°á»£c
            with open(output_filename, 'wb') as f:
                f.write(response.content)
            print(f"âœ… [THÃ€NH CÃ”NG] ÄÃ£ nháº­n vÃ  lÆ°u file Ã¢m thanh thÃ nh cÃ´ng.")
            print(f"   ğŸ’¾ File Ä‘Æ°á»£c lÆ°u táº¡i: {os.path.abspath(output_filename)}")
        else:
            print(f"âŒ [THáº¤T Báº I] Pháº£n há»“i khÃ´ng pháº£i lÃ  file audio/mpeg.")
            print(f"   Ná»™i dung pháº£n há»“i: {response.text}")

    except requests.exceptions.RequestException as e:
        print(f"âŒ [THáº¤T Báº I] ÄÃ£ xáº£y ra lá»—i khi gá»i API: {e}")
    except Exception as e:
        print(f"âŒ [THáº¤T Báº I] Lá»—i khÃ´ng xÃ¡c Ä‘á»‹nh: {e}")
    finally:
        print("--- Káº¿t thÃºc kiá»ƒm tra endpoint: /tts ---")
```
